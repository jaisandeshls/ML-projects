# customer classification using KNN

This classifier uses K-nearest neighbours to classify whether the customer had made a purchase or not.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('shopping.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
df['class'] = [1 if i == True else 0 for i in df['Revenue']]
```

```python
df.drop(columns=['Month','VisitorType','Weekend','Revenue'],inplace=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
features = df.drop(columns='class')
```

```python
X_train,X_test,y_train,y_test = train_test_split(features,df['class'],test_size=0.4)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=4)
```

```python
model.fit(X_train,y_train)
```

```python
predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(predict,y_test))
```

```python
print(classification_report(predict,y_test))
```
